## The if statement
Executes a block of code only if a condition is true.if the condition is false the entire block will be be skipped.

```python
if condition:
    # code to execute if condition is true
    # the block must be indented(4 spaces or 1 tab)
# code outside the if block run regardless
```

## The if-else statement
# syntax
if condition:
    # if condition is true
    # block of code will be execute
else:
    # code if the condition is false

```python
temperature = 30
if temperature >= 30:
    print(" it's a hot day")
else:
    print("it's not a hot day")
```

```python
temperature = 10
if temperature >= 30:
    print(" it's a hot day")
else:
    print("it's not a hot day")
```

```python
balance = 200
withdraw = 300
if withdraw >= balance:
    balance = balance - withdraw    
    print(f'withdraw : {balance}')
else:
    print('not sufficent')
    print(f'current balance is {balance}')
```
