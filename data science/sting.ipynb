## string
- A string is a sequence of character(letters,numbers,symbols) "data"
- It's used to represent text data in python


```python
# how too create a string
# By using single quote,double quote,triple quote
```

```python
# using single quote
name1 = "lasya"
# using double quotes
name2 = "lasya"
# using triple quotes
name = """lasya"""
# using triple quotes
name3 = '''lasya'''
print(name1)
print(type(name1))
print(name2)
print(type(name2))
print(name3)
print(type(name3))
```

## len()

```python
len(fruits)
```

```python
institute = 'number of students'
len(institute)
```
